**Import Statements**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn import svm
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score
```

**Task 1: Dataset Understanding & Basic Exploration**

```python

df = pd.read_csv('StudentsPerformance.csv')
print(df.head())
print("Shape: ",df.shape)
print("Column: ",df.columns)
df.info()
```

X- Independent Variables
- gender 
- race/ethnicity 
- parental level of education        
- lunch
- test preparation course
- leftover dependent variables

Y- Dependent Variables
- math score
- reading score
- writing score 

We will consider only writing score as dependent variable as it is linear regression and moreover we will use either math score or reading score as independent variables because in linear regression numerical values are only accepted

**Task 2: Data Preprocessing**


```python
print(df.isnull().sum())
X = df[['math score']]
y = df[['writing score']]
```

- Splitting Data (Training set-80% Testing set 20%)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

**Task 3: Implement Linear Regression Model**

- Initialisation and training of model

```python
reg = LinearRegression()
reg.fit(X_train,y_train)
```

- Printing of slope and intercept

```python
print("Slope: ",reg.coef_)
print("Intercept: ",reg.intercept_)
```

Coefficent basically represents the change in writing score wrt to change in every unit of math score (dy/dx)

**Task-4  Prediction & Visualization**

- Prediction

```python
y_predic = reg.predict(X_test)
```

- Plotting

```python
plt.scatter(X_test, y_test, color='blue', label='Actual Data') 
plt.plot(X_test, y_predic, color='red', label='Regression Line') 
plt.xlabel('Math Score')
plt.ylabel('Writing Score')
plt.legend()
plt.show()
```

**Task 5: Model Evaluation**

- Mean Absolute Error (MAE)

```python
print("MAE: ",mean_absolute_error(y_true=y_test,y_pred=y_predic))
```

- R2 Score

```python
print("R^2 Score: ",r2_score(y_true=y_test,y_pred=y_predic))
```

**Conclusion:**
- The model is performing reasonable because the R2 score is greater than 0.5, which means Math scores are a avg predictor of Writing scores. 
- MAE of 7.84 shows that our predictions are on average side not too far not too close to the actual data.
